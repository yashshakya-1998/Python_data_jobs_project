```python
# Importing libraries
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt

#Loading data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

#Data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
top_countries_salary = df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short', aggfunc='median')
```

```python
top_countries = df.value_counts('job_country').head(6).index
top_countries
```

```python
job_titles_top=df.value_counts('job_title_short').head(3).index
job_titles_top
```

```python
top_countries_salary=top_countries_salary[job_titles_top]
top_countries_salary = top_countries_salary.loc[top_countries]
top_countries_salary.plot(kind='bar')
plt.title('Which are the top countries in data science jobs')
plt.ylabel('Median salary ($USD)')
plt.legend(title='Job title')
plt.xlabel('')
plt.xticks(rotation=45, ha='right')
plt.show()
```
