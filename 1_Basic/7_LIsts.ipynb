```python
jobs = ['Data Analyst', 'Data Engineer', 'Data Scientist','Business Analyst']
# Option 1: Unpack all four items
job1, job2, job3, job4 = jobs
print(job1)
print(job2)
print(job3)
print(job4)

# Option 2: Use * to catch remaining items
# job1, job2, *others = jobs
```

```python
jobs = ['Data Analyst', 'Data Engineer', 'Data Scientist', 'Business Analyst', 'ML Engineer']

# Unpack the first two and catch the rest in 'others'
job1, job2, *others = jobs

print(f'Job 1: {job1}')
print(f'Job 2: {job2}')
print(f'Remaining jobs: {others}')
```
