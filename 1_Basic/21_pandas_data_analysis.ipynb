```python
from datasets import load_dataset

df = load_dataset("lukebarousse/data_jobs")
```

```python
import pandas as pd

# Convert the dataset split to a pandas DataFrame
df = df['train'].to_pandas()

# Now you can use .info()
df.info()
```

```python
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
# Converts the job_posted_month values from string to pandas datetime format
```

```python
df.value_counts('job_title_short')
```

```python
import pandas as pd

# Calculate salary statistics
salary_stats = df.groupby('job_title_short')['salary_year_avg'].agg(['min', 'max', 'median'])

# Calculate the total count of each job title
job_counts = df['job_title_short'].value_counts()

# Combine the results
result = salary_stats.join(job_counts.rename('total_postings'))
result = result.sort_values('total_postings', ascending=False)
result
```

```python
# To see the median salaries sorted by annual salary:
df = df.groupby('job_title_short')[['salary_year_avg', 'salary_hour_avg']].median()
df.sort_values('salary_year_avg', ascending=False)
```
