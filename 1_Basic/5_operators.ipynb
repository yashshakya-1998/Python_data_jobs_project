```python
salary = [100000,99000,150000,200000,920000,45000,75000,60000,60000]
avg_salary=sum(salary)/len(salary)
median_salary=sorted(salary)[len(salary)//2]
if avg_salary>median_salary:
    print(f"Avg salary: {avg_salary} is greater than Median salary: {median_salary}")
```
