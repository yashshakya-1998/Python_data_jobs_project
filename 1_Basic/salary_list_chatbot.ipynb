```python
# Sample dataset (15 records — under 20)
data = [
    {"job_title": "Data Analyst", "salary": 50000},
    {"job_title": "Data Engineer", "salary": 65000},
    {"job_title": "Data Scientist", "salary": 72000},
    {"job_title": "Data Analyst", "salary": 85000},
    {"job_title": "Data Engineer", "salary": 90000},

    {"job_title": "Data Engineer", "salary": 110000},
    {"job_title": "Data Analyst", "salary": 135000},
    {"job_title": "Data Scientist", "salary": 160000},
    {"job_title": "Data Engineer", "salary": 200000},
    {"job_title": "Data Scientist", "salary": 250000},

    {"job_title": "Data Scientist", "salary": 180000},
    {"job_title": "Data Analyst", "salary": 220000},
    {"job_title": "Data Engineer", "salary": 350000},
    {"job_title": "Data Scientist", "salary": 500000},
    {"job_title": "Data Analyst", "salary": 920000}
]
```

```python
salaries = [item["salary"] for item in data]
avg_salary=sum(salaries) / len(salaries)
median_salary=sorted(salaries)[len(salaries) // 2]
print(f"Average salary: {avg_salary}")
print(f"Median salary: {median_salary}")
```

```python
user_salary = 200000

if user_salary < 100000:
    print("Entry Level Salary Range")
elif 100000 <= user_salary < 300000:
    print("Mid Level Salary Range")
else:
    print("Senior / Executive Salary Range")
```

```python
from collections import defaultdict

# Group salaries by job title
job_salaries = defaultdict(list)
for item in data:
    job_salaries[item['job_title']].append(item['salary'])

print(f"{'Job Title':<20} | {'Average':<10} | {'Median':<10}")
print("-" * 45)

for title, salaries in job_salaries.items():
    avg = sum(salaries) / len(salaries)
    med = sorted(salaries)[len(salaries) // 2]
    print(f"{title:<20} | {avg:<10.2f} | {med:<10}")
```

```python
# Filter salaries for both roles
analyst_salaries = [item['salary'] for item in data if item['job_title'] == 'Data Analyst']
engineer_salaries = [item['salary'] for item in data if item['job_title'] == 'Data Engineer']

# Calculate averages
avg_analyst = sum(analyst_salaries) / len(analyst_salaries)
avg_engineer = sum(engineer_salaries) / len(engineer_salaries)

difference = avg_analyst - avg_engineer

print(f"Average Data Analyst Salary: ${avg_analyst:,.2f}")
print(f"Average Data Engineer Salary: ${avg_engineer:,.2f}")
print(f"Difference: ${abs(difference):,.2f} {'more' if difference > 0 else 'less'} for Data Analysts.")
```

Based on the dataset, we calculated the mean for each group and found the absolute difference between them.
