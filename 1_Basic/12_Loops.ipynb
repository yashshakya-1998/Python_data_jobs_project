### ***while***  Loops

```python
years_experience={
 'Alice': 5,
 'Bob': 12,
 'Charlie': 3,
 'David': 18,
 'Emma': 7,
 'Frank': 10,
 'Grace': 2,
 'Hannah': 15,
 'Ian': 6,
 'Julia': 9,
 'Kevin': 1,
 'Laura': 14,
 'Michael': 8,
 'Nina': 4,
 'Oliver': 16,
 'Paula': 11,
 'Quentin': 0,
 'Rachel': 13,
 'Sam': 19,
 'Tina': 5,
 'Umar': 17,
 'Victoria': 6,
 'William': 9,
 'Xavier': 3,
 'Zara': 12
}
years = 0
index = 0
years_list = list(years_experience.items())
while years < 10:
  key, years = years_list[index]
  if years >= 10:
    print(key, 'has', years, 'years of experience')
  index += 1
```
