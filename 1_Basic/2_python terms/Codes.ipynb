###Functions
A reusable piece of code that performs specific task.<br>
A fucntion starts with "def" and name_of_function().

```python
def greet():
  return "hello"
```

```python
greet()
```

```python
job_title = "Data Analyst"
job_location = "India"
job_salary = 120000
```

```python
print(f"JOB: {job_title}\nLOCATION: {job_location}\nSALARY: ${job_salary:,.0f}")
```

```python
def display_info(title, location, salary):
  return print(f"JOB: {title}\nLOCATION: {location}\nSALARY: ${salary:,.0f}")
```

```python
display_info(job_title, job_location, job_salary)
```

###Classes
A template for creating objects.<br>


```python
job_title = "Data Analyst"
job_location = "India"
job_salary = 120000
```

```python
class Job_postings:
  def __init__(self, title, location, salary):
    self.title = title
    self.location = location
    self.salary = salary
```

```python
Job_postings(job_title, job_location, job_salary)
```

###Attributes
A field in a record defined by its class.<br>

```python
class Job_postings:
  def __init__(self, title, location, salary):
    self.title = title
    self.location = location
    self.salary = salary
```

```python
job_1 = Job_postings(job_title, job_location, job_salary)
```

```python
print("JOB:", job_1.title)
print("Country:", job_1.location)
print(f"Salary: ${job_1.salary:,.0f}")
```

###Methods
These are basically the functons of an object.<br>

```python
class job_postings:
  def __init__(self, title, location, salary):
    self.title = title
    self.location = location
    self.salary = salary
  def display_info(self):
    return print(f"JOB: {self.title}\nLOCATION: {self.location}\nSALARY: ${self.salary:,.0f}")
```

```python
job_1 = job_postings(job_title, job_location, job_salary)
```

```python
job_1.display_info()
```
