```python
from datasets import load_dataset

df = load_dataset("lukebarousse/data_jobs")
```

```python
import pandas as pd

# Convert the dataset split to a pandas DataFrame
df = df['train'].to_pandas()

# Now you can use .info()
df.info()
```

```python
type(df.job_posted_date[0])
```

```python
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
# Converts the job_posted_month values from string to pandas datetime format
```

```python
df['job_posted_month'] = df.job_posted_date.dt.month_name()
# Creates a new column job_posted_month using the dt.month_name() function
# Also we can create no format of job_posted_month using the dt.month function
```

```python
df.sort_values(by='job_posted_date', inplace=True)
# Sorts the job_posted_date column default ascending:True, inplace means replce the new sorted volumn with the existing column.
```

```python
df.drop(labels='salary_hour_avg', axis=1, inplace=True)
# Drops the salary_hour_avg column.
```

```python
df.dropna(subset=['salary_year_avg'], inplace=True)
# Drops the rows with missing values in the specified column
```
